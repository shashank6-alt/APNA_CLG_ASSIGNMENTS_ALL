```python
class BankAccount:
    def __init__(self, account_number, owner_name, balance):
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited: {amount}. New balance: {self.balance}"
        else:
            return "Deposit amount must be positive."

    def withdraw(self, amount):
        if 0 < amount < self.balance :
            self.balance -= amount 
            return f"Withdrew: {amount}. New balance: {self.balance}"
        else :
            return "enter  a vallid amount accordin to your balance"

    def check_balance(self):
        return f"current balance : {self.balance}"


# Example usage:
input("Press Enter to create a bank account...")
account = BankAccount("123456789", "John Doe", 1000)
print(account.deposit(500))        # Deposited: 500. New balance: 1500
print(account.withdraw(200))       # Withdrew: 200. New balance: 1300
print(account.check_balance())     # Current balance: 1300   
```

```python
class student  :
    def __init__(self , __name , __roll_number , __marks) :
        self.__name = __name
        self.__roll_number = __roll_number
        self.__marks = __marks

    def get_name(self):
        if self.__name is None :
            return "Name is not defined"
        else :
            return self.__name
    def get_roll_number(self):
        if self.__roll_number not in range (0,100) :
            return "Roll number is not defined"
        return self.__roll_number
    def get_marks(self):
        if self.__marks >= 0:
            return self.__marks
        else:
            return "Marks must be non-negative"
        
student1 = student("Itadori", 99, 95)
print("Name:", student1.get_name())               # Output: Name: Alice
print("Roll Number:", student1.get_roll_number()) # Output: Roll Number: 99
print("Marks:", student1.get_marks())           # Output: Marks: 95

    
```

```python
class shape :
    def area(self):
        pass
class rectangle(shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
class circle(shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
class triangle(shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height
    
rect = rectangle(5, 10)
print("Rectangle area:", rect.area())  # Output: Rectangle area: 50
circle1 = circle(7)
print("Circle area:", circle1.area())    # Output: Circle area: 153.
tri = triangle(4, 8)
print("Triangle area:", tri.area())    # Output: Triangle area: 16.0

    
```

```python
class vehicle :
    def __init__(self, model , brand ):
        self.model = model
        self.brand = brand

class car(vehicle):
    def __init__(self, model, brand, num_seats):
        super().__init__(model, brand)
        self.num_seats = num_seats

    def display_info(self):
        return f"Car Model: {self.model}, Brand: {self.brand}, Number of Seats: {self.num_seats}"

class bike(vehicle):
    def __init__(self, model , brand , engine_cc ):
        super().__init__(model, brand)
        self.engine_cc = engine_cc
        
    def display_info(self):
        return f"Bike Model: {self.model}, Brand: {self.brand}, Engine CC: {self.engine_cc}"
        
car1 = car("Model S", "Tesla", 5)
print(car1.display_info())  # Output: Car Model: Model S, Brand: Tesla, Number of Seats: 5
bike1 = bike("Ninja", "Kawasaki", 600)
print(bike1.display_info())  # Output: Bike Model: Ninja, Brand: Kawasaki, Engine CC: 600
```

```python
from abc import ABC, abstractmethod

class employee(ABC):
    def __init__(self, name, position):
        self.name = name
        self.position = position

    @abstractmethod
    def salary(self):
        """Return the salary for this employee."""

class concrete_employee(employee):
    def salary(self):
        if self.position == "Intern":
            return 20000
        elif self.position == "FullTimeEmployee":
            return 80000
        elif self.position == "ContractEmployee":
            return 50000
        else:
            return 0

employee1 = concrete_employee("Alice", "FullTimeEmployee")
print("Your Salary:", employee1.salary())  # Output: Salary: 80000
employee2 = concrete_employee("Bob", "Intern")
print("Your Salary:", employee2.salary())    # Output: Salary: 20000
```

```python
class Person:
    def __init__(self, name, age=None, address=None):
        self.name = name
        self.age = age
        self.address = address

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Address:", self.address)
p1 = Person("Rahul")
p1.display()
p2 = Person("Anita", 28)
p2.display()    
p3 = Person("Vikram", 35, "123 Main St")
p3.display()    
```

```python
class Player:
    # class variable
    player_count = 0

    def __init__(self, name, level):
        # instance variables
        self.name = name
        self.level = level

        # increase player count whenever a player is created
        Player.player_count += 1

    def display(self):
        print("Name:", self.name)
        print("Level:", self.level)
    @classmethod
    def get_player_count(cls):
        return cls.player_count 
    
p1 = Player("Alice", 5)
p1.display()
print("Total Players:", Player.get_player_count())  # Output: Total Players: 1
```

```python
class Carnivorre :
    def hunt(self):
        return "Hunting for prey."
    
class Herbivorre :
    def graze(self):
        return "Grazing on plants."
    
class Omnivorre :
    def eat(self):
        return "Eating both plants and animals."
    
class Bear(Omnivorre, Herbivorre , Carnivorre):
    def roar(self):
        return "Roaring loudly!"
bear = Bear()
print(bear.roar())       # Output: Roaring loudly!
print(bear.eat())        # Output: Eating both plants and animals.
print(bear.graze())      # Output: Grazing on plants.
print(bear.hunt())       # Output: Hunting for prey.
```

```python
class User:
    def __init__(self, username):
        self.username = username
class Message:
    def __init__(self, sender, content):
        self.sender = sender
        self.content = content
class ChatRoom:
    def __init__(self):
        self.users = []
        self.messages = []
    def join(self, user):
        self.users.append(user)
        print(user.username, "joined the chatroom")
    def leave(self, user):
        self.users.remove(user)
        print(user.username, "left the chatroom")
    def send_message(self, sender, content):
        if sender in self.users:
            msg = Message(sender, content)
            self.messages.append(msg)
        else:
            print("User not in chatroom")

    def show_chat(self):
        print("\nChat History:")
        for msg in self.messages:
            print(msg.sender.username + ":", msg.content)

u1 = User("Rahul")
u2 = User("Anita")

room = ChatRoom()

room.join(u1)
room.join(u2)

room.send_message(u1, "Hello Anita!")
room.send_message(u2, "Hi Rahul!")

room.show_chat()

room.leave(u2)
```
