```python
print("Enter two numbers :")
num1 = int(input()) 
num2 = int(input())
sum = num1 + num2
print(f"The sum of {num1} and {num2} is {sum}")
sub = num1 - num2
print(f"The subtraction of {num1} and {num2} is {sub}")
product = num1 * num2
print(f"The product of {num1} and {num2} is {product}") 
quotient = num1 / num2
print(f"The quotient of {num1} and {num2} is {quotient}")
    
```

```python
print("enter radius of the circle:" )
r = float(input())
pi = 3.1416 
area = pi * r**2
circumference = 2 * pi * r  
print(f"The area of the circle is {area}")
print(f"The circumference of the circle is {circumference}")
```

```python
print("enter principal amount:")
P = float(input())
print("enter rate of interest:")
R = float(input())
print("enter time in years:")
T = float(input())
SI = (P * R * T) / 100
print(f"The simple interest is {SI}")
```

```python
print("Enter the decimal number :")
decimal_number = float(input())
num1 = int(decimal_number)
num2 = float(decimal_number)
num3 = num2 - num1
print(f"The integer value is {num1}, the float value is {num2} and the fractional part is {num3:.2f}")
```
