```python
def count_num():
    county = 0
    print("Enter the number to give its digits ")
    num = int(input())
    print("The number is :", num)
    while num > 0:
        digit = num % 10
        county = county + digit 
        num = num // 10
    return county
result = count_num()
print("The sum of digits is:", result)
```

```python
def calculator():
    print("Enter the first integer: ")
    a = int(input())
    print("second integer:")
    b = int(input())
    print("Enter the operator :")
    OP = input()
    if(OP == "+"):
        print("Adding...")
        print (a + b)
    elif(OP == "-"):
        print("Subtracting...")
        print(a-b)
    elif( OP == "*"):
        print("Multiplying...")
        print(a*b)
    elif(OP== "/"):
        print("Dividing...")
        print(a/b)

calculator()



```

```python
def is_prime():
    print("Enter the number to check if its prime or not:")
    num = int(input())
    if num <= 1:
        print("The number is not prime")
        return
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print("The number is not prime")
            return
    print("The number is prime")

is_prime()
```

```python
print("number guessing game")
import random
number_to_guess = random.randint(1, 100)
attempts = 0
while True:
    print("Enter your guess (between 1 and 100):")
    guess = int(input())
    attempts += 1
    if guess < number_to_guess:
        print("Too low! Try again.")
    elif guess > number_to_guess:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You've guessed the number {number_to_guess} in {attempts} attempts.")
        break

```
