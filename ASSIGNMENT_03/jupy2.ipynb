```python
print("Creating even and odd elements tuple from given input tuple")
tup_main =tuple(map(int, input("Enter the tuple elements separated by space :").split()))
even_tup = tuple()
odd_tup = tuple()
for el in tup_main:
    if el%2==0:
        even_tup += (el,)
    else:
        odd_tup += (el,)
print("Even elements tuple :", even_tup)
print("Odd elements tuple :", odd_tup)
```

```python
def dictionary_of_students():
    std_dict = {}
    std_name = list(map(str, input("Enter the student name :").split()))
    std_marks = list( input("Enter the student marks :").split())
    for i in range(len(std_name)):
        std_dict[std_name[i]] = std_marks[i]
    print("Enter any option to perform on the student dictionary :")
    print("A. Add Student\nB.Update marks \nC. Display Student and marks Dictionary\nD. Search for student \n")
    ran = input("Enter your choice (A/B/C/D) :").upper()
    if ran == 'A':
        new_name = input("Enter the new student name to add :")
        new_marks = int(input("Enter the new student marks to add :"))
        std_dict[new_name] = new_marks
        print("Student added successfully")
        print(std_dict)
    elif ran == 'B':
        upd_name = input("Enter the student name to update marks :")
        if upd_name in std_dict:
            upd_marks = int(input("Enter the new marks to update :"))
            std_dict[upd_name] = upd_marks
            print("Marks updated successfully")
            print(std_dict)
        else:
            print("Student not found in the dictionary")
    elif ran =='c':
        print("Student and Marks Dictionary :", std_dict)
    elif ran == 'D':
        search_name = input("Enter the student name to search :")
        if search_name in std_dict:
            print("Student found. Marks :", std_dict[search_name])
        else:
            print("Student not found in the dictionary")


print("start dictionary operations for students")

dictionary_of_students()            
```
